desta área para o usuário.

```
base = float(input("Qual o tamanho da base do quadrado: "))
height = base
area = base * height
double = area * 2
print("Á area do seu quadrado é: {} e o dobro dela é: {}".format(area, double))
```

6. Faça um Programa que peça 2 números inteiros e um número real. Calcule e
mostre:
  - o produto do dobro do primeiro com metade do segundo .
  - a soma do triplo do primeiro com o terceiro.
  - o terceiro elevado ao cubo.

```
numbers_str = input("Digite 3 números separados por espaço e sem pontuação: ")
numbers_list = [float(n) for n in numbers_str.split()]
one = numbers_list[0]
two = numbers_list[1]
three = numbers_list[2]

print("O produto do dobro do primeiro com metade do segundo é: {}".format(one * 2 + two / 2))
print("A soma do triplo do primeiro com o terceiro é: {}".format(one * 3 + three))
print("O terceiro elevado ao cubo é: {}".format(three ** 3))
```

8. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê: 
  - salário bruto. 
  - quanto pagou ao INSS. 
  - quanto pagou ao sindicato. 
  - o salário líquido. 
  - calcule os descontos 
  - salário líquido


```
hours = float(input("Qual o valor em R$ você ganhar por hora? "))
numbers_of_hours = float(input("Quantas horas você trabalhar por mês? "))
bruto = hours * numbers_of_hours
ir = bruto * 11 / 100
inss = bruto * 8 / 100
sindicato = bruto * 5 / 100
liquido = bruto - ir - inss - sindicato
print("\nO valor do seu salário bruto esse mês é: {}".format(bruto))
print("O valor do seu salário liquido esse mês é: {}".format(liquido))
print("\n== IMPOSTOS ==\n* IR: {}\n* INSS: {}\n* Sindicado: {}".format(ir, inss, sindicato))
```

9. Faça um programa que leia 2 strings e informe o conteúdo delas seguido do seu
comprimento. Informe também se as duas strings possuem o mesmo comprimento e
são iguais ou diferentes no conteúdo.

```
one = input("Digite uma sequência de caracteres: ")
two = input("Digite outra sequência de caracteres: ")
print("Tamanho de '{}' é: {} caracteres".format(one, len(one)))
print("Tamanho de '{}' é: {} caracteres".format(two, len(two)))
if (len(one) == len(two)):
  print("As duas strings são de tamanhos iguais!")
else:
  print("As duas strings são de tamanhos diferentes!")

if (one == two):
  print("As duas strings tem o mesmo conteúdo!")
else:
  print("As duas strings não tem o mesmo conteúdo")
```

11. Faça um programa que solicite a data de nascimento (dd/mm/aaaa) do usuário e imprima a data com o nome do mês por extenso. 

```
date = input("Digite a sua data de nascimento: ")
date = [int(x) for x in date.split("/")]
months = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembto', 'Outubro', 'Novembro', 'Dezembro']
print("Você nasceu em {} de {} de {}!".format(date[0], months[date[1]-1], date[2]))
```

12. Leet é uma forma de se escrever o alfabeto latino usando outros símbolos em
lugar das letras, como números por exemplo. A própria palavra leet admite muitas
variações, como l33t ou 1337. O uso do leet reflete uma subcultura relacionada ao
mundo dos jogos de computador e internet, sendo muito usada para confundir os
iniciantes e afirmar-se como parte de um grupo. Pesquise sobre as principais formas
de traduzir as letras. Depois, faça um programa que peça uma texto e transforme-o
para a grafia leet speak.

```
import re

text = input("Digite uma frase: ")
clone = text
text = re.sub("A|a", "4", text)
text = re.sub("T|t", "7", text)
text = re.sub("L|l", "1", text)
text = re.sub("E|e", "3", text)
text = re.sub("S|s", "5", text)
text = re.sub("O|o", "0", text)

print("Sua frase '{}' mudou para: {}".format(clone, text))
```
